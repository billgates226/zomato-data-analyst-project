```python
import pandas as pd
```

```python
df = pd.read_csv("C:\\Users\\gates\\Desktop\\data_analyst_mini_project\\zomato\\data\\zomato_bangalore_data.csv")
df.head()
```

Drop duplicates

```python
df.drop_duplicates(inplace=True)
```

"Columns like url, phone, and dish_liked were dropped as they were irrelevant to the core business analysis objectives."

```python
df.drop(['url', 'phone'], axis=1, inplace=True, errors='ignore')
```

```python
df.head()
```

Clean column names

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
```

```python
df['approx_cost(for_two_people)'] = df['approx_cost(for_two_people)'].str.replace(",", "").astype(float)
```

```python
df.head(1)
```

Clean rating column

```python
df['rate'] = df['rate'].str.split("/").str[0].str.strip()
df = df[df['rate'] != 'NEW']
df = df[df['rate'] != '-']
df['rate'] = df['rate'].astype(float)
```

Preview cleaned data

```python
print(df.head())
print(df.info())
```

```python
# save the cleaned data
df.to_csv("C:\\Users\\gates\\Desktop\\data_analyst_mini_project\\zomato\\data\\cleaned_zomato.csv", index=False)
```
