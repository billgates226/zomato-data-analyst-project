```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the cleaned data
df = pd.read_csv("C:\\Users\\gates\\Desktop\\data_analyst_mini_project\\zomato\\data\\cleaned_zomato.csv")

# Optional: Check the structure
print(df.head())
print(df.columns)
```

```python
df_cuisine = df['cuisines'].dropna().str.split(',').explode().str.strip().value_counts().head(10)
```

Top 10 Most Common Cuisines

```python
plt.figure(figsize=(10,6))
sns.barplot(x=df_cuisine.values, y=df_cuisine.index, color='yellow')  # Use single color
plt.title("Top 10 Most Common Cuisines")
plt.xlabel("Number of Restaurants")
plt.ylabel("Cuisine Type")
plt.savefig("C:\\Users\\gates\\Desktop\\data_analyst_mini_project\\zomato\\images\\top_10_cuisines.png", bbox_inches='tight')
plt.show()
```

Rating Distribution

```python
plt.figure(figsize=(8,5))
sns.histplot(df['rate'], bins=20, kde=True, color='red')
plt.title("Distribution of Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.savefig("C:\\Users\\gates\\Desktop\\data_analyst_mini_project\\zomato\\images\\Rating_distribution.png", bbox_inches='tight')
plt.show()
```

Top 10 Locations with Most Restaurants

```python
top_locations = df['location'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_locations.values, y=top_locations.index, palette='magma')
plt.title("Top 10 Locations with Most Restaurants")
plt.xlabel("Number of Restaurants")
plt.ylabel("Location")

plt.savefig("C:\\Users\\gates\\Desktop\\data_analyst_mini_project\\zomato\\images\\top_10locations_with_most_resturants.png", bbox_inches='tight')

plt.show()
```

Cost vs Rating

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='approx_cost(for_two_people)', y='rate', hue='online_order', alpha=0.6)
plt.title("Cost vs Rating (Colored by Online Order)")
plt.xlabel("Cost for Two People")
plt.ylabel("Rating")

plt.savefig("C:\\Users\\gates\\Desktop\\data_analyst_mini_project\\zomato\\images\\cost_vs_rating.png", bbox_inches='tight')

plt.show()
```

Online Delivery vs Ratings (Boxplot)

```python
plt.figure(figsize=(6,5))
sns.boxplot(data=df, x='online_order', y='rate', palette='Set2')
plt.title("Online Order Availability vs Rating")
plt.xlabel("Online Order")
plt.ylabel("Rating")

plt.savefig("C:\\Users\\gates\\Desktop\\data_analyst_mini_project\\zomato\\images\\online_devlivery_vs_rating.png", bbox_inches='tight')

plt.show()
```
